```python
x = np.arange(0, 5, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.axis('equal')
plt.show()
```

```python
x = np.linspace(-1, 1, 100)
y = np.sqrt(1 - x ** 2)
plt.plot(x, y)
plt.axis('scaled')
plt.show()
```

```python
t = np.linspace(0, 2 * np.pi, 100)
(x, y) = (np.cos(t), np.sin(t))
plt.plot(x, y)
plt.axis('scaled')
plt.show()
```

```python
y = np.arange(-100, 100, 0.01)
# noinspection PyShadowingNames
fn = lambda y: (-(2*a)**(-1)) * np.log(np.abs(y))
for a in [x for x in range(-5, 5) if x != 0]:
    plt.plot(fn(y), y)

plt.axis('scaled')
plt.ylim([-1, 1])
plt.xlim([-1, 1])
plt.grid()
plt.show()
```
